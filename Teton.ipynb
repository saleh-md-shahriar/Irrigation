```python
import numpy as np
import pandas as pd
```

```python
dfTrain = pd.read_csv('/content/TRAIN.csv').dropna()
dfTest = pd.read_csv('/content/Test.csv').dropna()
```

```python
dfTrain.head()
```

```python
dfTest.head()
```

```python
features = ['Humidity', 'Temperature']
```

```python
X_train = dfTrain[features]
y_train = dfTrain.Water
```

```python
X_test = dfTest[features]
y_test = dfTest.Water
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(2,)),
    keras.layers.Dense(4, activation=tf.nn.relu),
	keras.layers.Dense(4, activation=tf.nn.relu),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam',
              loss='mse',
              metrics=['mse'])

model.fit(X_train, y_train, epochs=300, batch_size=1)
```

```python
y_bar = model.predict(X_test)
```

```python
summation = 0  
n = 20 
for i in range (0,n): 
  difference = y_test[i] - y_bar[i]  
  squared_difference = difference**2  
  summation = summation + squared_difference  
MSE = summation/n 
print(MSE)
```

```python
model.save("model.h5")
```
